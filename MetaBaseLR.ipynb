```python
import requests

url = "http://127.0.0.1:3000/api/dataset"

payload = "{\"database\":1,\"type\":\"query\",\"query\":{\"source-table\":2},\"parameters\":[]}"
headers = {
    'content-type': "application/json",
    'cache-control': "no-cache",
    'X-Metabase-Session': "d245f925-ca4e-4448-9094-443d4e659918"
    }

response = requests.request("POST", url, data=payload, headers=headers)

print(response.text)
```

```python
lst=response.json()['data']['rows']
```

```python
df = pd.DataFrame(lst) 
```

```python
df.head()
```

```python
df=df.fillna(value=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(df.drop([1,6],axis=1), df[6])
```

```python
reg.score(df.drop([1,6],axis=1), df[6])
```
